```python
import requests
import json

OPENROUTER_API_KEY = "paste open router key"

url = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
    "Content-Type": "application/json",
}

def chat_with_bot(message):
    data = {
        "model": "openai/gpt-4o-mini",  # you can change model
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": message}

        ]
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))
    result = response.json()

    return result

# Simple chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    reply = chat_with_bot(user_input)
    print("Bot:", reply)
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=userdata.get("paste api key here")
)

doctor_prompt = """
You are a professional doctor conducting a patient interview.
Your goal is to gather medical history and understand the patient's current condition.
1. Start by asking for their name, age, and reason for visit.
2. Ask one question at a time about their symptoms, past medical history, medications, and lifestyle.
3. Provide empathetic, supportive feedback after each answer before moving to the next question.
4. Keep the conversation professional, compassionate, and clear.
5. At the end, summarize the key findings and highlight any concerning issues that need further medical evaluation.
6.limit of questions should be 2-5
"""

agent = create_agent(
    model=llm,
    tools=[],
    system_prompt=doctor_prompt,
    checkpointer=InMemorySaver(),
)
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
import uuid
from google.colab import userdata

# Initialize the LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",   # You can swap to gemini-1.5-flash if quota issues occur
    google_api_key=userdata.get("paste api key here")
)

# Placement interview prompt
placement_prompt = """
You are a placement officer interviewing about campus placements.
1. Ask for institution name and academic year.
2. Ask 2–4 questions about number of students placed, student reviews, and companies visited.
3. Give clear, professional feedback after each answer.
4. End with a concise summary of placements, reviews, and recruiters.
"""

# Create the agent
agent = create_agent(
    model=llm,
    tools=[],
    system_prompt=placement_prompt,
    checkpointer=InMemorySaver(),
)

# Interactive loop
config = {"configurable": {"thread_id": str(uuid.uuid4())}}

print("Placement Agent: Hello! I'm here to gather placement information.")

first_turn = True

while True:
    if first_turn:
        user_input = input("\nTo start, please enter your institution name and academic year: ")
        first_turn = False
    else:
        user_input = input("\nYour answer (or type 'End' to stop): ")

    if user_input.strip().lower() == "end":
        break

    result = agent.invoke(
        {"messages": [HumanMessage(content=user_input)]},
        config=config,
    )

    print(f"\nPlacement Agent: {result['messages'][-1].content}")

print("\nThank you for sharing placement details. This concludes the interview.")
```
